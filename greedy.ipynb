```python
def greedy(graph, color):
    V = len(graph)
    
    #dodeljujemo boju prvom cvoru    
    color[0] = 1
    num_of_colors = 1

    #dodeljujemo boje ostalim cvorovima
    for i in range(1, V):        
        neighbors = [index for index in graph[i]]
        used_colors = set([v for k, v in color.items() if k in neighbors])
        
        for j in range(1, V+1):
            if j not in used_colors:
                break
                
        # cvoru dodeljujemo prvu boju koja vec nije zauzeta
        color[i] = j

        if j > num_of_colors:
            num_of_colors = j
            
    return num_of_colors
```

```python
def verticesDegrees(graph):
    degrees = []
    V = len(graph)
    
    for i in range(V):
        degrees.append(len(graph[i]))
        
    return degrees
```

1. Izracunaj stepen svakog cvora
2. Sortiraj cvorove u opadajucem redosledu prema stepenu cvora
3. Dodeli boju 1 prvom cvoru
4. Prolazi kroz ostatak liste i dodeli istu tu boju svim cvorovima koji nisu povezani sa obojenim cvorom.
5. Ponavljaj korak 4 za sve neobojene cvorove sa novom bojom, u opadajucem redosledu dok nisu obojeni svi cvorovi.

```python
def WelshPowell(graph, color, degrees):
    #sortiramo cvorove opadajuce prema stepenu
    vertices = list(zip(*sorted([(deg, i) for i, deg in enumerate(degrees)], reverse=True, key=lambda x: (x[0], -x[1]))))[1]
    degrees.sort(reverse=True)
    
    clr_num = 0
    for i in range(len(degrees)):
        if vertices[i] not in color:
            clr_num = clr_num + 1
            color[vertices[i]] = clr_num
            #prolazimo kroz ostatak liste gde su cvorovi manjeg stepena
            for j in range(i+1, len(degrees)): 
                #ukoliko cvor nije obojen i nije susedan trenutnom cvoru => bojimo ga istom bojom
                if vertices[j] not in graph[vertices[i]] and vertices[j] not in color:
                    for k in range(j):
                        if vertices[k] in graph[vertices[j]] and vertices[k] in color:
                            if color[vertices[k]] == color[vertices[i]]:
                                break
                    else:
                        color[vertices[j]] = color[vertices[i]]
                    
    return clr_num
```

```python
def loadGraph(filename):
    with open(str(filename), "r") as graph_instance:
        chromatic_num = int(graph_instance.readline())
        V, E = list(map(int, graph_instance.readline().split(" ")))
        graph = [[] for _ in range(V)]
        for line in graph_instance:
            if line.startswith('e'):
                line = line.split(" ")
                del line[0]
                edge = list(map(int, line))
                if (edge[1]-1) not in graph[edge[0]-1]:
                    graph[edge[0] - 1].append(edge[1] - 1)
                    graph[edge[1] - 1].append(edge[0] - 1)
    return chromatic_num, graph
```

```python
def printSolution(color):
    print("Solution: ", end = " ")
    for k,v in sorted(color.items()):
        print(str(color[k]), end=" ")
        
    print()
```

```python
graph_list = ['brute.txt', 'myciel3.col', 'myciel4.col', 'myciel5.col', 'myciel6.col', 'david.col', 'huck.col', 'jean.col']
import time
from termcolor import colored

for graph_file in graph_list:
            print()
            print(colored(graph_file, color="red"))
            graph = []
            chromatic_num, graph = loadGraph(graph_file)
            color = {} #na pocetku ni jedan cvor nema dodeljenu boju, mapa cvor -> dodeljena boja
            degrees = verticesDegrees(graph)
                
            start = time.time()
            num_of_colors = WelshPowell(graph, color, degrees)
            end = time.time()
            exec_time = end - start
                
            print('time: ', exec_time)
            printSolution(color)
            print(f'Chromatic number: {chromatic_num}, Found chromatic number: {num_of_colors}')
```
